```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\tarun\OneDrive\Desktop\Project csv\Attrition_data.csv")
```

```python
df.head()
```

```python
print(df['Education'].unique())
```

```python
df['Education'] = df['Education'].astype('Int64')
df['JobSatisfaction'] = df['JobSatisfaction'].astype('Int64')
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].astype('Int64')
df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].astype('Int64')
df['JobInvolvement'] = df['JobInvolvement'].astype('Int64')
df['PerformanceRating'] = df['PerformanceRating'].astype('Int64')
df['WorkLifeBalance'] = df['WorkLifeBalance'].astype('Int64')
```

```python
education_map = {1: 'Below College', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor'}
satisfaction_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
rating_map = {1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding'}
worklife_map = {1: 'Bad', 2: 'Good', 3: 'Better', 4: 'Best'}
```

```python
df['Education'] = df['Education'].map(education_map)
df['JobSatisfaction'] = df['JobSatisfaction'].map(satisfaction_map)
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].map(satisfaction_map)
df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].map(satisfaction_map)
df['JobInvolvement'] = df['JobInvolvement'].map(satisfaction_map)
df['PerformanceRating'] = df['PerformanceRating'].map(rating_map)
df['WorkLifeBalance'] = df['WorkLifeBalance'].map(worklife_map)
```

```python
df.to_csv(r"C:\Users\tarun\OneDrive\Desktop\Project csv\Attrition_data_cleaned.csv", index=False)
print("Cleaned and saved!")
```
